4) Dado o valor de faturamento mensal de uma distribuidora, detalhado por estado:
• SP – R$67.836,43
• RJ – R$36.678,66
• MG – R$29.229,88
• ES – R$27.165,48
• Outros – R$19.849,53

Escreva um programa na linguagem que desejar onde calcule o percentual de representação que cada estado teve dentro do valor total mensal da distribuidora.  

```python
import matplotlib.pyplot as plt
```

```python
distribuidora = {'SP': 67_836.43, 'RJ': 36_678.66, 'MG': 29_229.88, 'ES': 27_165.48, 'OUTROS': 19_849.53}
espaco = [0.03, 0.01, 0.01, 0.01, 0.01]

cor = plt.get_cmap('Dark2').colors
plt.pie(distribuidora.values(), labels=distribuidora.keys(), colors=cor, explode=espaco, autopct='%.1f')
plt.title("Percentual de representação da distribuidora em cada estado")

plt.show
```
